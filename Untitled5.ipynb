###**Using scikit-image**

```python
import skimage.io as io
from skimage.transform import rescale
import numpy as np
```

```python
import numpy as np
import skimage.io as io
from skimage.transform import rescale
from skimage.color import rgb2gray

# Reading the image using its path
image = io.imread('scenery.jpeg')

# Rescale the image by a scaling factor of 0.7
rescaled_image = rescale(image, 0.7)

# Convert the image to a format supported by PIL
rescaled_image = (rescaled_image * 255).astype(np.uint8)
rescaled_image = rescaled_image[:, :, 0]

# Save the rescaled image with a filename
filename = 'savedImage.jpeg'
io.imsave(filename, rescaled_image)

# Display the rescaled image

io.imshow(rescaled_image)
io.show()
```

#**IMAGE AUGMENTATION**

```python
#importing all the required lib
import skimage.io as io
from skimage.transform import rotate
import numpy as np
import matplotlib.pyplot as plt
image=io.imread('scenery.jpeg')
def image_augment(image):
  fig, ax=plt.subplots(nrows=1,ncols=3,figsize=(9,8))
  ax[0].imshow(image)
  ax[0].axis('off')
  ax[1].imshow(rotate(image,angle=45,mode='wrap'))
  ax[1].axis('off')
  ax[2].imshow(np.fliplr(image))
  ax[2].axis('off')
image_augment(image)
```

## **EXPLORING IMAGES**

```python
#IMPORTING REQUIRED LIB
import skimage.io as io
#image=io.imread()
image=io.imread('scenery.jpeg')
import matplotlib.pyplot as plt
io.imshow(image)

#check the image matrix data type(could know the bit depth of the image)
print(image.dtype)
#check the height of image
print(image.shape[0])
#check the width of image
print(image.shape[1])
#check the number of channels of the image
print(image.shape[2])
```

##**Using PIL**

```python
from PIL import Image
im=Image.open('scenery.jpeg')
width,height=im.size
#display height and width
print(width)
print(height)
#print the filename
print(im.filename)
#print the format of the image
print(im.format)
#print the format description of the image
print(im.format_description)
```
